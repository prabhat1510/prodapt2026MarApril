1. What is Logging?

Logging means recording application events such as:

    errors
    warnings
    debugging information
    system status

Example events:

    Application started
    Database connected
    User login failed
    File not found

Logging helps with:

    ✔ Debugging
    ✔ Monitoring systems
    ✔ Production troubleshooting
    ✔ Auditing

2. Logging vs Print
Print	                         Logging
Used for temporary debugging	Used in production
No log levels	                Multiple log levels
Cannot save to file easily	    Can save to file
Not configurable	            Highly configurable

Example (Bad)
    print("Error connecting to database")

Example (Good)
    logging.error("Error connecting to database")
    
3. Python Logging Levels

Python provides 5 standard log levels.

    Level	Usage
    DEBUG	Detailed debugging information
    INFO	General system information
    WARNING	Something unexpected happened
    ERROR	Serious problem
    CRITICAL	System failure

Example

    import logging
    
    logging.debug("Debug message")
    logging.info("Info message")
    logging.warning("Warning message")
    logging.error("Error message")
    logging.critical("Critical error")

4. Basic Logging Example
    import logging
    
    logging.basicConfig(level=logging.INFO)
    
    logging.info("Application started")
    logging.warning("Low disk space")
    logging.error("File not found")

    Output
    
    INFO:root:Application started
    WARNING:root:Low disk space
    ERROR:root:File not found

5. Logging with Custom Format

Better logs include time + level + message.
    
    import logging
    
    logging.basicConfig(
        level=logging.INFO,
        format="%(asctime)s - %(levelname)s - %(message)s"
    )
    
    logging.info("Application started")
    logging.warning("CPU usage high")
    logging.error("Database connection failed")

   Output

    2026-03-16 10:10:10 - INFO - Application started
    2026-03-16 10:10:11 - WARNING - CPU usage high
    2026-03-16 10:10:12 - ERROR - Database connection failed

6. Logging to a File

    In real systems logs are written to files.

    import logging
    
    logging.basicConfig(
        filename="app.log",
        level=logging.INFO,
        format="%(asctime)s - %(levelname)s - %(message)s"
    )
    
    logging.info("Application started")
    logging.error("Something went wrong")

    Generated file app.log

        2026-03-16 10:12:10 - INFO - Application started
        2026-03-16 10:12:15 - ERROR - Something went wrong

7. Using Logger Object (Best Practice)

    Instead of logging.info(), create a logger object.

        import logging
        
        logger = logging.getLogger("MyApp")
        
        logging.basicConfig(level=logging.INFO)
        
        logger.info("Application started")
        logger.error("Payment failed")

8. Logging in Functions (Debugging Example)
    import logging
    
    logging.basicConfig(level=logging.DEBUG)
    
    def divide(a, b):
    
        logging.debug(f"Dividing {a} by {b}")
    
        try:
            result = a / b
            logging.info(f"Result = {result}")
            return result
    
        except ZeroDivisionError:
            logging.error("Division by zero error")
    
    divide(10,2)
    divide(10,0)

    Output
    
        DEBUG:root:Dividing 10 by 2
        INFO:root:Result = 5.0
        DEBUG:root:Dividing 10 by 0
        ERROR:root:Division by zero error
9. Structured Logging (Important for Production)

    Structured logs are logs written in JSON format.

    They are used in:
        Microservices
        Cloud systems
        ELK stack
        Splunk
        Datadog

    Example structured log

        {
         "timestamp":"2026-03-16",
         "level":"INFO",
         "service":"payment",
         "message":"Payment successful",
         "user_id":123
        }

10. Structured Logging Example in Python
        import logging
        import json
        
        logging.basicConfig(level=logging.INFO)
        
        def log_event(level, message, **kwargs):
        
            log_data = {
                "level": level,
                "message": message,
                "data": kwargs
            }
        
            log_message = json.dumps(log_data)
        
            if level == "INFO":
                logging.info(log_message)
            elif level == "ERROR":
                logging.error(log_message)
        
        log_event(
            "INFO",
            "User login",
            user_id=101,
            ip="192.168.1.10"
        )
        
        log_event(
            "ERROR",
            "Payment failed",
            user_id=101,
            amount=500
        )

    Output
    
    INFO:root:{"level":"INFO","message":"User login","data":{"user_id":101,"ip":"192.168.1.10"}}

11. Real Production Example (Data Pipeline)

    Example: CSV processing with logs
    
        import logging
        import csv
        
        logging.basicConfig(
            filename="pipeline.log",
            level=logging.INFO,
            format="%(asctime)s %(levelname)s %(message)s"
        )
        
        def process_file():
        
            logging.info("Starting CSV processing")
        
            try:
                with open("data.csv") as file:
        
                    reader = csv.reader(file)
        
                    for row in reader:
                        logging.debug(f"Processing row {row}")
        
                logging.info("File processed successfully")
        
            except Exception as e:
                logging.error(f"Processing failed: {e}")
        
        process_file()

     Log file

        INFO Starting CSV processing
        DEBUG Processing row ['101','John']
        INFO File processed successfully

12. Logging Best Practices (Industry)

        ✔ Never use print() in production
        ✔ Use correct log levels
        ✔ Use structured JSON logs in cloud systems
        ✔ Rotate log files
        ✔ Mask sensitive data (passwords)
        
        Example sensitive data masking:
        
        Password=****

13. Advanced Logging (Very Important for DevOps)

    Production systems use:
    
        Feature	Tool
        Log rotation	RotatingFileHandler
        Centralized logs	ELK Stack
        Cloud logs	AWS CloudWatch
        Monitoring	Prometheus + Grafana
    
    Example log rotation

        from logging.handlers import RotatingFileHandler
        import logging
        
        handler = RotatingFileHandler(
            "app.log",
            maxBytes=1000000,
            backupCount=3
        )
        
        logger = logging.getLogger()
        logger.setLevel(logging.INFO)
        logger.addHandler(handler)
        
        logger.info("Application started")
